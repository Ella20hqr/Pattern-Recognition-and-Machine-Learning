```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
y = 1
y_hat = np.arange(-1, 2, 0.1)
labels = ['LDA', 'Logistic Regression', 'SVM']
```

```python
lda = (y_hat - y)**2/2
lr = -np.log((np.exp(y_hat))/(1+np.exp(y_hat)))
hinge = []
for i in y_hat:
    if 1-i>0:
        hinge.append(1-i)
    else:
        hinge.append(0)
# hinge = np.max(1-y_hat, 0)
```

```python
fig, ax = plt.subplots()
ax.plot(y_hat, hinge, label='SVM')
ax.plot(y_hat, lr, label='Logistic Regression')
ax.plot(y_hat, lda, label='LDA')
ax.legend()
ax.set_ylabel('Single-data Loss Function')
ax.set_xlabel(r'Predicted values:$w^{\top} x_i + b$')
# ax.axhline(1, linestyle='--', color='grey')
plt.savefig('loss_func.png', dpi=100)
plt.show()
```
