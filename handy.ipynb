```python
## load data
data = np.genfromtxt('hw2_linear_model/breast-cancer-wisconsin.txt', dtype=int)
# the np.genfromtxt load '?' as -1
# print('Missing values\' index:')
# print(np.array(np.where(data==-1)).T)
# all the missing values are in the 7th column
# fill them with the mean of the 7th column
data[np.where(data==-1)] = np.mean(data[:, 6])

## split the features and labels
X = data[:,1:10]
y = data[:, -1]
ind_good = y==1
ind_bad = y==0
```

```python
for i in range(9):
    plt.hist(X[ind_good, i], bins=10, color='r', alpha=0.5)
    plt.hist(X[ind_bad, i], bins=10, color='g', alpha=0.5)
    plt.title(i)
    plt.show()
```
